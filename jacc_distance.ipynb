```python
df_filtered.columns = gene_ranks.columns
```

```python
def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return len(s1.intersection(s2)) / len(s1.union(s2))
```

```python
ncols = len(df_filtered.columns)
jacc_sim = np.zeros((ncols, ncols))
```

```python
for i in range(ncols):
    print("Working with column: ", i)
    for j in range(ncols):
        l1 = list(df_filtered.iloc[:,i][df_filtered.iloc[:,i] != 0].index)
        l2 = list(df_filtered.iloc[:,j][df_filtered.iloc[:,j] != 0].index)
        jacc_sim[i][j] = jaccard_similarity(l1, l2)
```

```python
jacc_sim[0:15][0:15]
```

```python
sns.heatmap(jacc_sim)
```

```python
jacc_sim_df = pd.DataFrame(jacc_sim, index = df_filtered.columns, columns = df_filtered.columns )
```

```python
jacc_sim_df.head()
```

```python
links = []
for i in range(ncols):
    for j in range(i+1, ncols):
        if(jacc_sim[i][j] > 0.30):
            links.append([jacc_sim_df.columns[i], jacc_sim_df.columns[j], jacc_sim[i][j]] )
```
