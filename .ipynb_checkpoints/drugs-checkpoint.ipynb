```python
import pandas as pd
```

```python
drug=pd.read_csv("data/cell_lines_drugs.csv")
gene=open("data/gene.titles","r").read()
gene=gene.split("\n")
gene.remove("")
for i in range(len(gene)):
    gene[i]=gene[i].split(" (")[0]
    print(gene[i])
```

```python
dic={}
for i in range(len(drug[drug.columns[1]])):
    dic[drug[drug.columns[1]][i]]=drug[drug.columns[2]][i]
```

```python
LIST={}
cnt=0
for i in dic:
    tmp=str(dic[i]).split(", ")
    if tmp!="nan":
        for g in tmp:
            if g in gene:
                cnt+=1
                if i not in LIST.keys():
                    LIST[i]=[g]
                else:
                    LIST[i].append(g)
print("found ", cnt, "corresponding genes")        
```

```python
print(len(LIST.keys()), "corresponding drugs")
```

```python
ww=open("data/listofdrugs.csv","w")
ww.write("drug_name,genes\n")
for i in LIST.keys():
    ww.write(i+",|")
    for j in LIST[i]:
        ww.write(j+"|")
    ww.write("\n")
ww.close()
```
